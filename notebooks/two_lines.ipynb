```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_linear_equations(a11, a12, b1, a21, a22, b2):
    # Define the x range for plotting
    x = np.linspace(-10, 10, 400)
    
    # Calculate y values for each equation
    y1 = (b1 - a11 * x) / a12
    y2 = (b2 - a21 * x) / a22
    
    # Calculate the intersection point
    A = np.array([[a11, a12], [a21, a22]])
    B = np.array([b1, b2])
    intersection = np.linalg.solve(A, B)
    
    # Plot the equations
    plt.figure(figsize=(8, 6))
    plt.plot(x, y1, label=f'{a11}x + {a12}y = {b1}')
    plt.plot(x, y2, label=f'{a21}x + {a22}y = {b2}')
    
    # Plot the intersection point
    plt.plot(intersection[0], intersection[1], 'ro', label='Intersection Point')
    plt.text(intersection[0], intersection[1], f'  ({intersection[0]:.2f}, {intersection[1]:.2f})', fontsize=12)
    
    # Set plot limits
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    
    # Add labels and legend
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot of Linear Equations and Intersection Point')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    
    # Show plot
    plt.show()

    return
```

```python
# Example usage:
plot_linear_equations(1, -5, 8, 1, 1, 4)
```

```python
plot_linear_equations(1, -5, 8, 2, -4, 12)
```

# Visualization of higher dimentional spaces

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x1 and x2 values
x1 = np.linspace(-10, 10, 100)  # Change range as needed
x2 = np.linspace(-10, 10, 100)  # Change range as needed
x1, x2 = np.meshgrid(x1, x2)

# Calculate y values based on the plane equation
y = 2 * x1 + 4 * x2 + 5

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x1, x2, y, cmap='viridis', edgecolor='none')

# Add labels
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('3D Plane: y = 2*x1 + 4*x2 + 5')

# Show the plot
plt.show()
```

```python
from sympy import symbols, Eq, solve, Rational

# Definir las incógnitas
x, y, z = symbols('x y z')
```

```python
# Definir las ecuaciones
eq1 = Eq(x + y + z, 6)         
eq2 = Eq(2*x - y + 3*z, 14)    
eq3 = Eq(-3*x + 4*y + 2*z, -2)     

# Resolver el sistema de ecuaciones
solucion = solve((eq1, eq2, eq3), (x, y, z))

# Convertir los resultados a fracciones usando Rational
x_sol = Rational(solucion[x])
y_sol = Rational(solucion[y])
z_sol = Rational(solucion[z])

# Mostrar el resultado en fracciones
print(f"La solución es x = {x_sol} y = {y_sol} z = {z_sol}")
```
