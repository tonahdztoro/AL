```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Examen_2_df.xlsx')
```

```python
df['Original'] = df['Resultado']*100/15
df['Punto extra'] = (df['Resultado'] + 1)*100/15
```

```python
df
```

```python
# Set up the plot
plt.figure(figsize=(8, 6))

# Plotting the KDE for both columns with shaded areas
sns.kdeplot(df['Original'], bw_adjust=1, fill=True, label='Original', color='orange')
sns.kdeplot(df['Punto extra'], bw_adjust=1, fill=True, label='Punto extra', color='blue')

# Set the x-axis limits to display only from 50 to 100
plt.xlim(50, 120)
plt.axvline(x=100, color='green', linestyle='--', linewidth=2, label='100')

# Add titles and labels
plt.title('Distribuciones de los resultados')
plt.xlabel('Calificación')
plt.ylabel('Densidad')

# Add a legend to distinguish the two distributions
plt.legend()

# Show the plot
plt.show()
```
