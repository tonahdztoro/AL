```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Examen_df.xlsx')
```

```python
df
```

```python
# Count the occurrences of 0, 1, 2 in each column
value_counts = df.apply(pd.Series.value_counts).T

# Plotting the value counts for each column
value_counts.plot(kind='bar', figsize=(10, 6))

# Adding labels and title
plt.title('Distribution of Values (0, 1, 2) in Each Column')
plt.xlabel('Columns')
plt.ylabel('Count of Values')
plt.xticks(rotation=0)

# Show plot
plt.show()
```

```python
import numpy as np
```

```python
# Count occurrences of 0, 1, 2 in each column
value_counts = df.apply(pd.Series.value_counts).T

# Separate counts of 0s, 1s, and 2s
count_0 = value_counts[0]  # Count of 0s
count_1 = value_counts[1]  # Count of 1s
count_2 = value_counts[2]  # Count of 2s

# Create a new figure
fig, ax = plt.subplots(figsize=(10, 6))

# Define the index for bars (x positions)
indices = np.arange(len(value_counts.index))

# Bar width
bar_width = 0.35

# First bar: for column 1 (count of 0s)
ax.bar(indices, count_0, bar_width, label='0', color='orange')

# Second bar: stacked bars for columns 2 and 3 (count of 1s and 2s)
ax.bar(indices + bar_width, count_1, bar_width, bottom=count_2, label='1', color='deepskyblue')
ax.bar(indices + bar_width, count_2, bar_width, label='2', color='green')

# Adding labels and title
ax.set_title('Bar Plot with Separate 0 and Stacked 1, 2')
ax.set_xlabel('Columns')
ax.set_ylabel('Count of Values')

# Set x-ticks to be in the middle of the grouped bars
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(value_counts.index.astype(str))

# Adding legend
ax.legend()

# Show plot
plt.show()
```
